### User input for purchase

```python
ls = []

while (1 > 0):
    user_prod = input("Enter Product ID : ")
    quantity = int(input("Enter Quantity : "))
    
    for i in mI:
        
        if (user_prod == i):
            
            mI[user_prod]['Quantity'] = mI[user_prod]['Quantity'] - quantity
            print( user_prod ,"  ",mI[user_prod]['Product Name'],"  ", quantity,"  ",(quantity * mI[user_prod]['Price']))
    ls.append([user_prod , mI[user_prod]['Product Name'] , quantity, (quantity * mI[user_prod]['Price'])])
    check = input(" Want to enter another product : ")
    if (check == 'y'):
        continue
    elif (check == 'n'):
          break
```

### Bill For user

```python
print("****************************")
print("Product ID"," ","Product Name"," ","Quantity"," ","Price","\n")
for i in ls:
    print(i[0]," ",mI[i[0]]['Product Name']," ", quantity," ", (quantity) * mI[i[0]]['Price'])
print("****************************")

sum_T = 0

for i in ls:
    sum_T = sum_T + (quantity) * mI[i[0]]['Price']
print("Sum Total = ", sum_T)

if (sum_T > 5000):
    sum_T = sum_T - (sum_T * 0.05)
    print("Yay! You got a discount of 5 %!")
    print("Final amount = ", sum_T)

if (sum_T > 8000):
    sum_T = sum_T - (sum_T * 0.11)
    print("Yay! You got a discount of 11 %!")
    print("Final amount = ", sum_T)
print("****************************")
```
